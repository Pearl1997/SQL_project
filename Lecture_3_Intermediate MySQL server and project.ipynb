### Introducing MySQL database

We will use Sakila database for our SQL learning tasks.

The __Sakila__ sample database was initially developed by Mike Hillyer, a former member of the MySQL AB documentation team, and is intended to provide a standard schema that can be used for examples in books, tutorials, articles, samples, and so forth. Sakila sample database also serves to highlight the latest features of MySQL such as Views, Stored Procedures, and Triggers.

The Sakila database is a nicely normalised schema modelling a DVD rental store, featuring things like films, actors, film-actor relationships, and a central inventory table that connects films, stores, and rentals.


If your Mysql does not have Sakila, please follow the [instructions](https://dev.mysql.com/doc/sakila/en/sakila-installation.html) to install it.

#### Regex

__What is regex?__ 

A regular expression, regex or regexp (sometimes called a rational expression) is a sequence of characters that define a search pattern. Usually such patterns are used by string searching algorithms for "find" or "find and replace" operations on strings, or for input validation. It is a technique developed in theoretical computer science and formal language theory.

Regular expressions are used in search engines, search and replace dialogs of word processors and text editors, in text processing utilities such as sed and AWK and in lexical analysis. Many programming languages provide regex capabilities either built-in or via libraries.

__Why is it important to learn regex?__

* Functions are limited. Sometimes, regex is the only way to match the target from random text. 
* An easy tool to parse and extract information from free text. 
* Commonly used in almost all the programming languages.
* Good for your technical interviews. 

__How to learn regex?__ 

* [Online tool](https://rubular.com/)
* Don't need to remember everything. Bring a cheatsheet. 
* Practice regex in multiple programming languages, e.g. SQL, R, Python, SAS, etc.
